```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
# Inicializar modelos
lr_model = LinearRegression()
mlp_model = MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)

# Diccionarios para almacenar mtricas
lr_metrics = {'R2': [], 'RMSE': [], 'MAE': []}
mlp_metrics = {'R2': [], 'RMSE': [], 'MAE': []}

# Realizar validacin cruzada
for train_index, test_index in kf.split(X_scaled):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Entrenar y evaluar LinearRegression
    lr_model.fit(X_train, y_train)
    y_pred_lr = lr_model.predict(X_test)
    lr_metrics['R2'].append(r2_score(y_test, y_pred_lr))
    lr_metrics['RMSE'].append(np.sqrt(mean_squared_error(y_test, y_pred_lr)))
    lr_metrics['MAE'].append(mean_absolute_error(y_test, y_pred_lr))
    
    # Entrenar y evaluar MLPRegressor
    mlp_model.fit(X_train, y_train)
    y_pred_mlp = mlp_model.predict(X_test)
    mlp_metrics['R2'].append(r2_score(y_test, y_pred_mlp))
    mlp_metrics['RMSE'].append(np.sqrt(mean_squared_error(y_test, y_pred_mlp)))
    mlp_metrics['MAE'].append(mean_absolute_error(y_test, y_pred_mlp))

# Calcular estadsticos para cada mtrica
def compute_stats(metrics):
    stats = {}
    for metric, values in metrics.items():
        stats[metric] = {
            'Mean': np.mean(values),
            'Std': np.std(values),
            'Variance': np.var(values)
        }
    return stats

# Reportar resultados
lr_stats = compute_stats(lr_metrics)
mlp_stats = compute_stats(mlp_metrics)

print("\nLinearRegression Results (5-Fold CV):")
for metric, stats in lr_stats.items():
    print(f"{metric}: Mean={stats['Mean']:.4f}, Std={stats['Std']:.4f}, Variance={stats['Variance']:.4f}")

print("\nMLPRegressor Results (5-Fold CV):")
for metric, stats in mlp_stats.items():
    print(f"{metric}: Mean={stats['Mean']:.4f}, Std={stats['Std']:.4f}, Variance={stats['Variance']:.4f}")
```
