```python
import pandas as pd
df = pd.read_csv('auto-mpg.csv')
df.info()
```

```python
# Verificar valores nulos
df.isnull().sum()
```

```python
#Identificar y convertir valores "?" a NaN
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
print(df['horsepower'].isna().sum()) 
```

```python
# Imputar 'hoursepower' con la mediana
median_horsepower = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(median_horsepower)
df.isnull().sum()
```

```python
# Eliminar columna 'car name' (no aporta valor predictivo)
df = df.drop('car name', axis=1)
df.head()
```
