```python
import pandas as pd
df = pd.read_csv("auto-mpg.csv")
```

```python
import numpy as np
#Tratamiento de Valores Faltantes
df['horsepower'] = df['horsepower'].replace('?', np.nan)
df['horsepower'] = pd.to_numeric(df['horsepower'])
```

```python
# tratar nulos con la mediana
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
# Eliminar la columna 'car name'
df = df.drop('car name', axis=1)

X = df.drop('mpg', axis=1)
y = df['mpg']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Matriz de correlaci√≥n
corr_matrix = df.corr()
print("Correlaci√≥n con 'mpg':\n", corr_matrix['mpg'].sort_values(ascending=False))
# Visualizar matriz (opcional)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
# 2. RFE con regresi√≥n lineal
model = LinearRegression()
rfe = RFE(model, n_features_to_select=6) 
rfe_fit = rfe.fit(X, y)
rfe_ranking = pd.DataFrame({'Feature': X.columns, 'Ranking': rfe.ranking_, 'Selected': rfe.support_})
print("\nRFE Ranking:\n", rfe_ranking.sort_values(by='Ranking'))
```

```python
from sklearn.tree import DecisionTreeRegressor
# 3. √Årbol de decisi√≥n 
tree = DecisionTreeRegressor(random_state=42)
tree.fit(X, y)
tree_importance = pd.DataFrame({'Feature': X.columns, 'Importance': tree.feature_importances_})
print("\nImportancia √Årbol de Decisi√≥n:\n", tree_importance.sort_values(by='Importance', ascending=False))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split, cross_val_score
import numpy as np

selected_features = ['displacement', 'weight', 'model year', 'horsepower']

X_train, X_test, y_train, y_test = train_test_split(
    X[selected_features], 
    y, 
    test_size=0.2, 
    random_state=42
)
```

```python
# Modelo 1: Random Forest con K-Fold Cross-Validation
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf_scores_r2 = cross_val_score(rf, X[selected_features], y, cv=5, scoring='r2')
rf_scores_mae = cross_val_score(rf, X[selected_features], y, cv=5, scoring='neg_mean_absolute_error')
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Evaluacion del Random Forest")
print("CV R¬≤: Media =", np.mean(rf_scores_r2), "Desviaci√≥n=", np.std(rf_scores_r2), "Varianza =", np.var(rf_scores_r2))
print("CV MAE: Media =", -np.mean(rf_scores_mae), "Desviaci√≥n=", np.std(-rf_scores_mae), "Varianza =", np.var(-rf_scores_mae))
print("Test R¬≤:", r2_score(y_test, y_pred_rf))
print("Test MAE:", mean_absolute_error(y_test, y_pred_rf))
```

```python
# Modelo 2: MLP ( una red neuronal ) con K-Fold Cross-Validation
mlp = MLPRegressor(hidden_layer_sizes=(50,), max_iter=1000, random_state=42)
mlp_scores_r2 = cross_val_score(mlp, X[selected_features], y, cv=5, scoring='r2')
mlp_scores_mae = cross_val_score(mlp, X[selected_features], y, cv=5, scoring='neg_mean_absolute_error')
mlp.fit(X_train, y_train)
y_pred_mlp = mlp.predict(X_test)
print("Evaluacion del MLPRegressor")
print("CV R¬≤: Media =", np.mean(mlp_scores_r2), "Desviaci√≥n=", np.std(mlp_scores_r2), "Varianza =", np.var(mlp_scores_r2))
print("CV MAE: Media =", -np.mean(mlp_scores_mae), "Desviaci√≥n=", np.std(-mlp_scores_mae), "Varianza =", np.var(-mlp_scores_mae))
print("Test R¬≤:", r2_score(y_test, y_pred_mlp))
print("Test MAE:", mean_absolute_error(y_test, y_pred_mlp)) 
```

```python
# Gr√°ficas de dispersi√≥n
plt.figure(figsize=(12, 5))
# Dispersi√≥n Random Forest
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred_rf, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Valores Reales (mpg)')
plt.ylabel('Predicciones (mpg)')
plt.title('Random Forest: Predicciones vs Reales')
plt.grid(True)

# Dispersi√≥n MLP
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_mlp, color='green', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Valores Reales (mpg)')
plt.ylabel('Predicciones (mpg)')
plt.title('MLP: Predicciones vs Reales')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Gr√°ficas de residuos
plt.figure(figsize=(12, 5))

# Residuos Random Forest
residuals_rf = y_test - y_pred_rf
plt.subplot(1, 2, 1)
plt.scatter(y_pred_rf, residuals_rf, color='blue', alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicciones (mpg)')
plt.ylabel('Residuos')
plt.title('Random Forest: Gr√°fica de Residuos')
plt.grid(True)

# Residuos MLP
residuals_mlp = y_test - y_pred_mlp
plt.subplot(1, 2, 2)
plt.scatter(y_pred_mlp, residuals_mlp, color='green', alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicciones (mpg)')
plt.ylabel('Residuos')
plt.title('MLP: Gr√°fica de Residuos')
plt.grid(True)

plt.tight_layout()
plt.show()
```
